```
import pandas as pd
import numpy as np
import requests
import seaborn as sns
import matplotlib.pyplot as plt
import io
```

```
url="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
s=requests.get(url).content
titanic=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```
titanic.info()
```

```
titanic.rename(columns={'pclass': 'clase'}, inplace=True)
```

```
titanic.rename(columns={'fare': 'tarifa'}, inplace=True)
```

```
titanic.clase.unique()
```

```
titanic.head(5)
```

```
#Consulta con condiciones
titanic[
 (titanic.sex == 'female')
 & (titanic['clase'].isin(['1','2']))
 & (titanic.age > 35 )
 & (titanic.survived == 0)
]
```

```
sns.distplot (titanic.age.dropna( ))
sns.plt.show ( )
```

```
sns.set(font_scale=0.5)
g = sns.FacetGrid(titanic, row='survived',col='clase')
g.map(sns.distplot, "age")
sns.plt.show()
```

```
#Diagrama de dispersion con Distribucion de cada variable: tarifa(precio)/age(edad)
sns.jointplot(data=titanic, x='age', y='tarifa',
kind='reg', color='b')
sns.plt.show()
```

```
#Define un subconjuto de datos con las variables num√©ricas
titanic_num = titanic[['survived','clase','sibsp','parch','tarifa']]
#Hace una matriz de diagramas de dispersi√≥n de parejas de variables.
sns.pairplot(titanic_num, hue="survived")
sns.plt.show()
```

```
sns.set(style="ticks")
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species")
```

```
# A continuacion se visualiza la distribucion de clase/edad y los supervivientes del titanic
# Se usa la clase FacetGrid la cual ayuda visualizando la distribucion de una variable tambien con las relaciones entre multiples variables usando paneles, se usan 3 
# dimensiones: row (para la variable sex), col (para la variable clase) y hue (para la variable survived), las 2 primeras variables tienen correspondencia con los ejes
 
g = sns.FacetGrid(titanic, col="clase", row="sex", hue="survived",margin_titles=True)
g.map(plt.hist, "age")
g.add_legend();
```

```
# Se observa claramente que los supervivientes del titanic estaban muy directamente relacionados a mayor clase (1) y con su genero (female) y con una edad promedio de 23-27 a√±os
```
